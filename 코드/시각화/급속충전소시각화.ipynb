```python
import pandas as pd
import numpy as np
import folium
```

```python
data = pd.read_csv('../../데이터/test.csv')

data = data[data['충전소명'] != '충전소명']

latitude = data['위도'].astype(float)
longitude = data['경도'].astype(float)

data.columns
data['급속/완속'] = np.nan
data.loc[data['충전기타입'] == '02','급속/완속'] = '완속'
data.loc[data['충전기타입'] != '02', '급속/완속'] = '급속'
data['급속/완속']
# 01: DC차데모 (급속)
# 02: AC완속 (완속)
# 03: DC차데모+AC3상 (급속)
# 04: DC콤보 (급속)
# 05: DC차데모+DC콤보 (급속)
# 06: DC차데모+AC3상+DC콤보 (급속)
# 07: AC3상 (급속)
data
```

```python
m = folium.Map(location=[37.5,127], zoom_start=12, tiles = 'OpenStreetMap')
tooltip = '클릭해 주세요!'
k = 1
for i, j in zip(data['위도'].astype(float), data['경도'].astype(float)):
    if data.loc[k,'급속/완속'] == '급속':
        folium.Marker([i, j], popup='<button>ㅇㅇ</button>', tooltip=tooltip).add_to(m)
    
    k += 1
```
