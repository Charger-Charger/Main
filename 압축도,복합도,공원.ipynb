```python
import pandas as pd
import geopandas as gpd
import shapely

import pydeck as pdk
from shapely.geometry import Polygon, Point
```

```python
compress = gpd.read_file('데이터/compress/35.1 토지이용(건물) 압축도.shp', encoding = 'utf-8')

compress
```

```python
compress = compress[compress['sido_nm_k'] == '서울특별시']

compress['gid']
```

```python
compress = compress[['gid','value']]
compress = compress.set_index('gid')
compress.columns = ['토지이용_압축도']
compress
compress.to_excel('C:\data/토지이용압축도.xlsx')
```
